## SQL запрос: ##

### Получите результирующее множество из БД в соответствии с заданными условиями выборки, сортировки, группировки данных. Создайте отчеты с агрегацией данных. ###
```sql
-- Вывести всех пользователей, которые имеют больше одного члена семьи
SELECT *
FROM insurance_final_rows
WHERE family_members > 1
```
```sql
-- Посчитать среднюю зарплату среди женщин (gender = false)
SELECT AVG(salary)
FROM insurance_final_rows
WHERE gender = false
```
```sql
-- Найти возраст самого молодого пользователя, который обращался за страховой выплатой хотя бы один раз
SELECT MIN(age)
FROM insurance_final_rows
WHERE insurance_appeal > 0
```
```sql
-- Найти пользователя с максимальным количеством обращений за страховой выплатой и вывести его user_id, возраст и количество обращений
SELECT user_id, age, insurance_appeal
FROM insurance_final_rows
WHERE insurance_appeal = (SELECT MAX(insurance_appeal) FROM insurance_final_rows);
```

## Python: ##

````python
import pandas as pd
````

````python
# Загружаем CSV в DataFrame
df = pd.read_csv('insurance_final_rows.csv')
````

````python
# Находим пользователей с более чем одним членом семьи
users_more_than_one_family_member = df.loc[df['family_members'] > 1]
print("Пользователи с более чем 1 членом семьи:")
users_more_than_one_family_member
````

````python
# Средняя зарплата среди женщин (gender == False)
avg_salary_women = df.loc[df['gender'] == False, 'salary'].mean()
f'Средняя зарплата среди женщин: {avg_salary_women:.2f}'
````

````python
# Минимальный возраст среди обратившихся хотя бы один раз за страховкой
min_age_appealed = df.loc[df['insurance_appeal'] > 0, 'age'].min()
f"Возраст самого молодого пользователя с обращением: {min_age_appealed}"
````

````python
# Находим пользователя с максимальным количеством обращений за страховкой, выводим его user_id, возраст, кол-во обращений
max_appeals = df['insurance_appeal'].max()
user_max_appeals = df[df['insurance_appeal'] == max_appeals][['user_id', 'age', 'insurance_appeal']]
print("Пользователь с максимальным количеством обращений:")
user_max_appeals
````
